# NumPy 

NumPy (ou Numpy) é uma biblioteca de álgebra linear para Python, a razão pela qual é tão importante para a Data Science com Python é que quase todas as bibliotecas dependem do NumPy como um dos seus principais blocos de construção.

Numpy também é incrivelmente rápido, pois tem ligações para bibliotecas C. Para obter mais informações sobre por que você deseja usar Arrays em vez de listas, confira esta excelente publicação do [StackOverflow post](http://stackoverflow.com/questions/993984/why-numpy-instead-of-python-lists).

## Numpy Arrays

As matrizes de NumPy são a maneira principal de usar Numpy ao longo do curso. Numpy arrays essencialmente vêm de duas formas: vetores e matrizes. Os vetores são estritamente arranjos de 1d e as matrizes são 2d (mas você deve observar que uma matriz ainda pode ter apenas uma linha ou uma coluna).

Vamos começar nossa introdução explorando como criar matrizes numPy.

## Indexar e acessar dados de um array

Agora, aprenderemos como **indexar e acessar** dados de um array. Uma das maneiras mais simples de fazer isso é semelhante à indexação em uma lista.

```python
my_array = np.arange(0,11)
my_array
```

```python
my_array[8]
```

Também podemos acessar uma **fatia** do array (slice)

A sintaxe é:

*   a[start:stop]  # items start through stop-1
*   a[start:]      # items start through the rest of the array
*   a[:stop]       # items from the beginning through stop-1
*   a[:]           # a copy of the whole array
*   a[start:-stop]    # items start through (the size of the array - stop)

```python
my_array[2:7]
```

```python
my_array[5:]
```

```python
my_array[:5]
```

```python
my_array[:]
```

```python
my_array[5:-3]
```

Há também o valor do passo (step), que pode ser usado com qualquer um dos itens acima:
*   a[start:stop:step] # start through not past stop, by step

```python
my_array[1:7:3]
```

Até agora, mostramos as operações NumPy Arrays que são muito semelhantes às operações de lista, mas, um benefício (muito) de usar NumPy Arrays é que podemos **transmitir** dados

```python
my_array[0:5]=87
my_array
```

```python
my_array = np.arange(0,11)
arr_slice = my_array[0:6]
arr_slice
```

```python
arr_slice[:] = 2008
arr_slice
```

```python
my_array
```

```python
my_array_copy = my_array[0:6].copy()
my_array_copy
```

```python
my_array_copy[:] = 2009
my_array_copy
```

```python
my_array
```

## Reshape
Retorna uma matriz contendo os mesmos dados com uma nova forma.

```python
my_array = np.arange(12)
print(f"Array before re-shape\n {my_array}, \n shape:{my_array.shape}")
my_array_2 = my_array.reshape(4,3)
print(f"Array after re-shape\n {my_array_2}, \n shape:{my_array_2.shape}")

print(f"Array's size before re-shape {my_array.size},  Array's size before re-shape:{my_array_2.size}")
```

## Operaçoes aritimeticas

### Executando operações básicas com arrays usando NumPy

```python
# Create two arrays
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([6, 7, 8, 9, 10])
```

```python
# Add two arrays
add_arr = arr1 + arr2
add_arr
```

```python
# Subtract two arrays
sub_arr = arr2 - arr1
sub_arr
```

```python
# Multiply two arrays
mul_arr = arr1 * arr2
mul_arr
```

```python
# Divide two arrays
div_arr = arr2 / arr1
div_arr
```

```python
# Exponentiation
exp_arr = arr1 ** arr2
exp_arr
```

```python
# Modulo
mod_arr = arr1 % arr2
mod_arr
```

### Executando operações básicas com matrizes usando NumPy

```python
# Create two matrices
mat1 = np.array([[1, 2], [3, 4]])
mat2 = np.array([[5, 6], [7, 8]])
```

```python
# Addition
add_mat = mat1 + mat2
add_mat
```

```python
# Subtraction
sub_mat = mat1 - mat2
sub_mat
```

```python
# Element-wise multiplication
mul_mat = np.multiply(mat1, mat2)
mul_mat
```

```python
# Matrix multiplication
dot_mat = np.dot(mat1, mat2)
dot_mat
```

matrix_mul.png

```python
# Scalar multiplication
scalar_mat = 2 * mat1
scalar_mat
```

```python
# Transpose
trans_mat = np.transpose(mat1)
trans_mat
```

# Matplotlib

Matplotlib é uma biblioteca abrangente para criar visualizações estáticas, animadas e interativas em Python. O Matplotlib torna as coisas fáceis fáceis e as difíceis possíveis.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(start=0, stop=10, step=1)
y1 = np.random.rand(10)
y2 = np.random.rand(10)
print(x)
print(y1)
```

```python
plt.figure(figsize=(7,5))
plt.plot(x, y1, marker='o', c='blue')
plt.plot(x, y2, marker='x', c='red')
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.scatter(x, y1, c='red', marker='x')
plt.scatter(x, y2, c='green',s=10)
plt.show()
```

```python
plt.figure(figsize=(8,6))

x = list(range(10))

for idx in range(5):
  y = np.random.rand(10,1)
  plt.plot(x, y, label=f'Line {idx}')

plt.legend()
plt.show()
```

```python
x = np.arange(10)
y1 = np.random.rand(10)
```

```python
plt.figure(figsize=(7, 5))
plt.bar(x, y1, label='Bar plot')
plt.legend()
plt.show()
```

```python
x = np.arange(0, 2*np.pi, 0.05)
y1 = np.cos(x)
y2 = np.sin(x)

plt.figure(figsize=(8,6))
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

O padrão OpenCV para imagens coloridas é BGR, enquanto matplotlib usa RGB. Podemos usar a função interna de conversão de espaço de cores.

```python
img2 = cv.cvtColor(img2,code=cv.COLOR_BGR2RGB)
img2.shape
```

```python
img3 = cv.imread('image_processing_tools/images/intro/animal.jpg')
```

```python
img3 = cv.cvtColor(img3,code=cv.COLOR_BGR2RGB)
img3.shape
```

```python
plt.imshow(img3)
plt.show()
```

We can add two images by using function add. This directly adds up image pixels in the two images. 

```python
new_image = cv.add(img,img3)
```

```python
plt.imshow(new_image)
plt.show()
```

We can also add the images with weights to each image using **addWeighted**

```python
weight_img_1 = 0.5
weight_img_2 = 0.5
gamma = 0 #
```

```python
new_image = cv.addWeighted(img,weight_img_1,img3,weight_img_2,gamma)
```

```python
plt.imshow(new_image)
plt.show()
```

We can also subtract the image using the function subtract.

Image **Scaling**: Refers to the resizing of an image. It changes the number of pixels from an image, in-zooming or out-zooming.

```python
half = cv.resize(img, (0, 0), fx = 0.1, fy = 0.1)
bigger = cv.resize(img, (1000, 1000))
 
stretch_near = cv.resize(img, (780, 540),
               interpolation = cv.INTER_NEAREST)
 
 
Titles =["Original", "Half", "Bigger", "Interpolation Nearest"]
images =[img, half, bigger, stretch_near]
count = 4
plt.figure(figsize=(8, 6), dpi=80)
for i in range(count):
    plt.subplot(2, 2, i + 1)
    plt.title(Titles[i],y=-0.001,color='r')
    plt.imshow(images[i])
 
plt.show()
```
